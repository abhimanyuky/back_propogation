## Basic image filteration

#Import required packages

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

#Read the image

```python
image = cv2.imread("horse.jpg")
plt.imshow(image)
```

## Applying filter for 3x3 Kernel

```python
kernel_3x3 = np.ones((3,3), np.float32) / 9
blurred = cv2.filter2D(image, -1, kernel_3x3)
plt.imshow(blurred)
```

## Applying blur of 7x7 kernel

```python
kernel_7x7 = np.ones((7,7), np.float32) / 49

blurred2 = cv2.filter2D(image, -1, kernel_7x7)
plt.imshow(blurred2)
```

```python
canny_img = cv2.Canny(image,250,500)
plt.imshow(canny_img)
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
plt.imshow(gray_image)
```

```python
a = plt.hist(image.ravel(),256,[0,256])
```

```python
ret1,th1 = cv2.threshold(image,110,250,cv2.THRESH_BINARY)
```
